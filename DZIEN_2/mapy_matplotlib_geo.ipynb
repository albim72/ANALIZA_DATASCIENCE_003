```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
plt.figure(figsize=(12, 6))
m = Basemap(projection="robin",lon_0=0)

m.drawcoastlines()
m.drawcounties()
m.fillcontinents(color="lightgrey",lake_color="lightblue")
m.drawmapboundary(fill_color="lightblue")

plt.title("Mapy świata")
plt.show()
```

```python
#tworzenie mapy
fig,ax = plt.subplots(figsize=(12, 6), subplot_kw=dict(projection=ccrs.PlateCarree()))
ax.set_global()

#dodawanie cech geograficznych
ax.add_feature(cfeature.LAND, facecolor="lightgrey")
ax.add_feature(cfeature.OCEAN, facecolor="lightblue")
ax.add_feature(cfeature.COASTLINE, edgecolor="black")
ax.add_feature(cfeature.BORDERS, edgecolor="black")

plt.title("mapa świata")
plt.show()
```

```python
#tworzenie mapy
fig,ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(projection=ccrs.PlateCarree()))
ax.set_extent([14, 25, 49, 55])

#dodawanie cech geograficznych
ax.add_feature(cfeature.LAND, facecolor="lightgrey")
ax.add_feature(cfeature.OCEAN, facecolor="lightblue")
ax.add_feature(cfeature.COASTLINE, edgecolor="black")
ax.add_feature(cfeature.BORDERS, edgecolor="black")

plt.title("mapa Polski")
plt.show()
```

pobierz Shapefile województw z:

    Geoportal GUGiK
    Dane GUS (BDOT)

Po pobraniu Shapefile (np. plik ZIP), rozpakuj go i znajdź plik .shp zawierający województwa.

```python
import geopandas as gpd
from shapely.geometry import Polygon
import os

# Tworzenie przykładowych wielokątów (umowne granice województw)
wojewodztwa_data = {
    "nazwa": [
        "Dolnośląskie", "Kujawsko-Pomorskie", "Lubelskie", "Lubuskie",
        "Łódzkie", "Małopolskie", "Mazowieckie", "Opolskie", "Podkarpackie",
        "Podlaskie", "Pomorskie", "Śląskie", "Świętokrzyskie", "Warmińsko-Mazurskie",
        "Wielkopolskie", "Zachodniopomorskie"
    ],
    "geometry": [
        Polygon([(14, 50), (16, 50), (16, 52), (14, 52), (14, 50)]),  # Dolnośląskie
        Polygon([(16, 52), (18, 52), (18, 54), (16, 54), (16, 52)]),  # Kujawsko-Pomorskie
        Polygon([(22, 50), (24, 50), (24, 52), (22, 52), (22, 50)]),  # Lubelskie
        Polygon([(14, 52), (16, 52), (16, 54), (14, 54), (14, 52)]),  # Lubuskie
        Polygon([(18, 50), (20, 50), (20, 52), (18, 52), (18, 50)]),  # Łódzkie
        Polygon([(19, 49), (21, 49), (21, 50), (19, 50), (19, 49)]),  # Małopolskie
        Polygon([(20, 51), (22, 51), (22, 54), (20, 54), (20, 51)]),  # Mazowieckie
        Polygon([(16, 50), (18, 50), (18, 52), (16, 52), (16, 50)]),  # Opolskie
        Polygon([(21, 49), (23, 49), (23, 50), (21, 50), (21, 49)]),  # Podkarpackie
        Polygon([(22, 52), (24, 52), (24, 54), (22, 54), (22, 52)]),  # Podlaskie
        Polygon([(16, 54), (18, 54), (18, 56), (16, 56), (16, 54)]),  # Pomorskie
        Polygon([(18, 49), (20, 49), (20, 50), (18, 50), (18, 49)]),  # Śląskie
        Polygon([(20, 50), (22, 50), (22, 51), (20, 51), (20, 50)]),  # Świętokrzyskie
        Polygon([(18, 54), (20, 54), (20, 56), (18, 56), (18, 54)]),  # Warmińsko-Mazurskie
        Polygon([(18, 52), (20, 52), (20, 54), (18, 54), (18, 52)]),  # Wielkopolskie
        Polygon([(14, 54), (16, 54), (16, 56), (14, 56), (14, 54)]),  # Zachodniopomorskie
    ]
}

# Tworzenie GeoDataFrame
wojewodztwa_gdf = gpd.GeoDataFrame(wojewodztwa_data, crs="EPSG:4326")

# Zapis do pliku SHP
output_dir = "shapefiles"
os.makedirs(output_dir, exist_ok=True)  # Tworzenie folderu, jeśli nie istnieje
shp_path = os.path.join(output_dir, "wojewodztwa.shp")
wojewodztwa_gdf.to_file(shp_path)

print(f"Plik SHP zapisany do: {shp_path}")
```

```python
import matplotlib.pyplot as plt

# Wczytanie pliku SHP
wojewodztwa = gpd.read_file(shp_path)

# Rysowanie mapy
fig, ax = plt.subplots(figsize=(10, 8))
wojewodztwa.plot(ax=ax, column="nazwa", cmap="tab20", edgecolor="black", legend=True)

plt.title("Przykładowa mapa Polski – Województwa")
plt.axis("off")
plt.show()
```
