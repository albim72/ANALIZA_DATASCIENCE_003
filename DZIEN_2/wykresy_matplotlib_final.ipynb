```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from cycler import cycler
import numpy as np
```

```python
#wczytywanie danych z CSV
file_path = "dane.csv"
data = pd.read_csv(file_path)
```

```python
data
```

```python
data['Data'] = pd.to_datetime(data['Data'])
```

```python
#tworzenie wykresu liniowego na podstawie danych z ramki (matplotlib)
plt.figure(figsize=(10,6))
plt.plot(data['Data'],data['Wartosc'],marker='o',color='blue',label='Wartosc')
plt.scatter()
plt.title("Wykres liniowy wartości w czasie",fontsize=16)
plt.xlabel("Data",fontsize=12)
plt.ylabel("Wartośc",fontsize=12)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.legend()

plt.xticks(rotation=45)
plt.tight_layout()


plt.savefig("wykres.png")
plt.savefig("wykres.pdf")
plt.savefig("wykres.jpg")

plt.show()
```

```python
#wykres z wieloma słupkami
#przygotowanie danych
osoby = ["Anna","Piotr","Olga","Roman"]
informacje = {
    'wiek':(34,56,23,31),
    'waga':(56.8,89.6,60.3,105.6),
    'wzrost':(162,172,170,189)
}
```

```python
x = np.arange(len(osoby))
width = 0.25
multiplier = 0


fig,ax = plt.subplots(constrained_layout=True)
for attribute,mesurement in informacje.items():
    offset = width*multiplier
    rects = ax.bar(x+offset,mesurement,width,label=attribute)
    ax.bar_label(rects,padding=5)
    multiplier += 1

ax.set_ylabel('cecha')
ax.set_title('wybrane informacje na temat osób')
ax.set_xticks(x+width,osoby)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0,220)
plt.savefig("osoby.jpg")
plt.show()
```
